```python
import pandas as pd
```

```python
import os 
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/Border_Crossing_Entry_Data.csv")
```

```python
df = pd.DataFrame(data)
```

```python
state = df.groupby("State").sum()
state_df = state.sort_values(by="Value",ascending=True).drop(columns="Port Code")
```

```python
stateTotal = state_df[state_df["Value"]>700]
```

```python
stateTotal.plot(kind = "bar")
plt.ylabel("Per Billion")
plt.title("Border Crossings Per State")
plt.show()
```

```python
port = df.groupby("Port Name").sum()
port_df = port.sort_values(by="Value",ascending=True).drop(columns="Port Code")
```

```python
portTotal = port_df[port_df["Value"]> 23301176]
```

```python
portTotal.plot(kind="bar")
```

```python
northOrder = df[df["Border"] == "US-Canada Border"].reset_index()
```

```python
northBorder = northOrder.drop(columns=["index","Measure","Port Code"])
```

```python
southOrder = df[df["Border"] == "US-Mexico Border"].reset_index()
```

```python
southBorder = southOrder.drop(columns=["index","Measure","Port Code"])
```

```python
southStates = southBorder.groupby("State").sum()
```

```python
northStates = northBorder.groupby("State").sum()
```

```python
southStates.plot(kind = "bar")
plt.ylabel("Per Billion")
plt.title("Border Crossings Per State")
plt.show()
```

```python
northStates.plot(kind = "bar")
plt.ylabel("Per million")
plt.title("Border Crossings Per State")
plt.show()
```
